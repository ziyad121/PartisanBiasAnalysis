```python
import pandas as pd 
import matplotlib.pyplot as plt
df = pd.read_csv("data.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
#Number of records on each side:
df['label'].value_counts()
```

```python
#The newspapers included: 
df['Source'].value_counts().plot("bar", title="The newspapers included", figsize= (12,12) )
```

```python
#Sentiment analysis results:
print(df['flag'].value_counts())
df['flag'].value_counts().plot("bar", title="Total Tweets Sentiment", figsize= (7,7))
```

```python
#sentiment distribution: 
'''     positive sentiment: compound score >= 0.05
        neutral sentiment: (compound score > -0.05) and (compound score < 0.05)
        negative sentiment: compound score <= -0.05
'''
df.hist("compound_average", figsize= (10,10))
```

```python
# Positive reaction when trump's picture is present in the articles for both parties :

df[(df['trump_image']=="trump") & (df['flag']== "positive")]["label"].value_counts().plot("bar", title="Positive reaction on the presence of trump's picture", figsize= (7,7))
plt.show()

# Negative reaction when trump's picture is present in the articles for both parties :

df[(df['trump_image']=="trump") & (df['flag']== "negative")]["label"].value_counts().plot("bar", title="Negative reaction on the presence of trump's picture",figsize= (7,7))
plt.show()

# Negative reaction when trump's picture is present in the articles for both parties :

df[(df['trump_image']=="trump") & (df['flag']== "neutral")]["label"].value_counts().plot("bar", title="Neutral reaction on the presence of trump's picture",figsize= (7,7))
plt.show()
```

```python
# Positive reaction when trump's picture is absent in the articles for both parties :

df[(df['trump_image']!="trump") & (df['flag']== "positive")]["label"].value_counts().plot("bar", title="Positive reaction on the absence of trump's picture", figsize= (7,7))
plt.show()

# Negative reaction when trump's picture is absent in the articles for both parties :

df[(df['trump_image']!="trump") & (df['flag']== "negative")]["label"].value_counts().plot("bar", title="Negative reaction on the absence of trump's picture",figsize= (7,7))
plt.show()

# Neutral reaction when trump's picture is absent in the articles for both parties :

df[(df['trump_image']!="trump") & (df['flag']== "neutral")]["label"].value_counts().plot("bar", title="Neutral reaction on the absence of trump's picture",figsize= (7,7))
plt.show()
```
